###### 1.1 Latitude y Longitude a coordenadas cartesianas.

```python
listings_full['x']=np.cos(listings_full.latitude)*np.cos(listings_full.longitude)
listings_full['y']=np.cos(listings_full.latitude)*np.sin(listings_full.longitude)
listings_full['z']=np.sin(listings_full.latitude)

listings_full.head()
```

###### 1.2 Property_type.

```python
listings_full['property_type'].value_counts()
```

```python
listings_full.head()
```

```python
listings_full['property_type'] = listings_full['property_type'].str.replace(r'.*Serviced apartment.*', 'Apartment', regex=True)\
                                             .str.replace(r'.*Bed and breakfast.*', 'Hotel', regex=True)\
                                             .str.replace(r'.*Hut.*', 'Other', regex=True)\
                                             .str.replace(r'.*Pension.*', 'Other', regex=True)\
                                             .str.replace(r'.*Earth house.*', 'Other', regex=True)\
                                             .str.replace(r'.*Farm stay.*', 'Other', regex=True)\
                                             .str.replace(r'.*Barn.*', 'Other', regex=True)\
                                             .str.replace(r'.*Tent.*', 'Other', regex=True)\
                                             .str.replace(r'.*Bungalow.*', 'Other', regex=True)\
                                             .str.replace(r'.*Castle.*', 'Other', regex=True)\
                                             .str.replace(r'.*Nature lodge.*', 'Other', regex=True)\
                                             .str.replace(r'.*Igloo.*', 'Other', regex=True)\
                                             .str.replace(r'.*Casa particular (Cuba).*', 'Casa', regex=True)
```

```python
listings_full['property_type'].value_counts()
```

```python
property_type=['property_type']

listings_full=pd.get_dummies(listings_full, columns=property_type, drop_first=True)
```

###### 1.5. amenities

```python
listings_full. amenities.unique()
```

```python
amenities_full = set()
for amenities_list in listings_full.amenities.str.strip("{}").str.replace('"', '').str.split(','):
    amenities_full.update([amenity.strip() for amenity in amenities_list if amenity.strip() != '' and 'translation missing' not in amenity])

```

```python
amenities_full
```

```python
amenities_m = []
for fila in amenities:
    row = []
    for servicio in amenities_full:
        if servicio in fila:
            row.append(1)
        else:
            row.append(0)
    amenities_m.append(row)
```

```python
amenities = pd.DataFrame(amenities_m, columns=list(amenities_full))
```

```python
amenities.head()
```

```python
listings_full.drop('amenities', axis=1, inplace=True)
```

```python
listings_full=pd.concat([listings_full, amenities], axis=1)
```

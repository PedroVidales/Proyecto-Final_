```python
import os

filename = '../archive/Boxing.gzip'

if os.path.splitext(filename)[1] not in ['.gzip', '.gz']:
    raise ValueError("El archivo no estÃ¡ comprimido con gzip")
```

```python
import gzip
import pandas as pd

filename = '../archive/Boxing.gzip'

with gzip.open(filename, 'rb') as f:
    try:
        df = pd.read_csv(f, delimiter=',')
    except pd.errors.EmptyDataError:
        raise ValueError("El archivo no es un archivo CSV vÃ¡lido")
```

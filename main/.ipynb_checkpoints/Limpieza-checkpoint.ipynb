###### 3.2. Separaremos las coordenadas de la columna attraction_coordinates para manipularlas mejor después.

```python
import json


def separar_coordenadas(coordenadas):
    lista = json.loads(coordenadas)
    latitudes, longitudes = zip(*lista)
    return pd.Series({'latitudes': latitudes, 'longitudes': longitudes})
```

```python
ngdf[['latitudes', 'longitudes']] = ngdf['attraction_coordinates'].apply(lambda x: separar_coordenadas(x))
ngdf.head()
```

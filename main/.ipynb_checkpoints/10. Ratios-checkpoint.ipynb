```python
import folium

# encontrar los 1000 primeros alojamientos con mejor ratio positivo
top_300_positive = merged_df.nlargest(300, "ratio_positivo")



positive_indexes = top_300_positive.index
top_300_negative = merged_df[~merged_df.index.isin(positive_indexes)].nlargest(300, "ratio_negativo")
```

```python
top_300_positive_locations = top_300_positive[["latitude", "longitude"]]
top_300_negative_locations = top_300_negative[["latitude", "longitude"]]
```

```python
# Crear el mapa
m = folium.Map(location=[merged_df["latitude"].iloc[0], merged_df["longitude"].iloc[0]], zoom_start=12, width='70%', height='90%')

# Crear las capas de marcadores
positive_layer = folium.FeatureGroup(name='Alojamientos con mejor ratio positivo')
negative_layer = folium.FeatureGroup(name='Alojamientos con mejor ratio negativo')

# Agregar los marcadores a las capas correspondientes
for location in top_300_positive_locations.values:
    folium.Marker(location=location, icon=folium.Icon(color="green", icon_size=8)).add_to(positive_layer)

for location in top_300_negative_locations.values:
    folium.Marker(location=location, icon=folium.Icon(color="red", icon_size=8)).add_to(negative_layer)

# Agregar las capas al mapa
positive_layer.add_to(m)
negative_layer.add_to(m)

# Agregar el controlador de capas
folium.LayerControl().add_to(m)

# Mostrar el mapa
m
```

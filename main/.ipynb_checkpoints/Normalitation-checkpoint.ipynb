```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import geopandas as gpd
import seaborn as sns
import warnings
import regex as re

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
listings_done = pd.read_csv('../madrid/listings_detailed/listings_done.csv', low_memory=False)
listings_done.head()
```

```python
columns_norm = ['accommodates', 'bathrooms', 'bedrooms', 'beds', 'security_deposit', 'cleaning_fee', 'guests_included', 'extra_people', 'minimum_nights', 'maximum_nights', 'availability_30', 'availability_60', 'availability_90', 'availability_365','number_of_reviews','calculated_host_listings_count','calculated_host_listings_count_entire_homes','calculated_host_listings_count_private_rooms','calculated_host_listings_count_shared_rooms']
```

```python
listings_normalitation=listings_done.copy()
from scipy.stats import boxcox
```

```python
listings_normalitation[columns_norm]=boxcox(listings_normalitation[columns_norm], lmbda=0.618033)

listings_normalitation.head()
```

```python
df_norm = (listings_normalitation - listings_normalitation.mean()) / listings_normalitation.std()

# Verificamos que la media y la desviación estándar de los datos normalizados sean cercanas a 0 y 1, respectivamente
print('Media de los datos normalizados:', np.mean(df_norm))
print('Desviación estándar de los datos normalizados:', np.std(df_norm))
```

```python

plt.boxplot(listings_normalitation[columns_norm].values)


plt.xticks(range(1, len(columns_norm) + 1), columns_norm, rotation=90)

plt.title('Distribución de las variables normalizadas')
plt.xlabel('Variables')
plt.ylabel('Valores normalizados')

plt.show()
```

```python
listings_normalitation.to_csv('../madrid/listings_detailed/listings_normalitation.csv', index=False)
```

```python
noches_por_id = cal.groupby('listing_id')['minimum_nights'].sum()

# Calcular el dinero total ganado por cada listing_id
dinero_por_id = noches_por_id * cal.groupby('listing_id')['price'].mean()

# Encontrar el listing_id que más dinero ha ganado
id_mas_dinero = dinero_por_id.idxmax()

print(f"El id que más dinero ha ganado es: {id_mas_dinero}")
```

```python
import matplotlib.pyplot as plt

ingresos_por_id = noches_por_id * cal.groupby('listing_id')['price'].mean()

# Ordenar los listing_id por ingresos totales
ingresos_por_id = ingresos_por_id.sort_values()

# Seleccionar los primeros y últimos 10 listing_id
top10 = ingresos_por_id.tail(10)
bottom10 = ingresos_por_id.head(10)

# Calcular los logaritmos de los ingresos
log_top10 = np.log10(top10)
log_bottom10 = np.log10(bottom10)

# Crear la figura y el eje
fig, ax = plt.subplots()

# Trazar un gráfico de barras horizontal con los logaritmos de los ingresos
ax.barh(log_bottom10.index, log_bottom10, label='Menores ingresos')
ax.barh(log_top10.index, log_top10, label='Mayores ingresos')

# Añadir etiquetas y títulos
ax.set_xlabel('Logaritmo de ingresos totales')
ax.set_ylabel('Listing ID')
ax.set_title('Ingresos totales por Listing ID')
ax.legend()

# Mostrar el gráfico
plt.show()
```

```python
import folium

# Crear el mapa centrado en la ciudad de Madrid
madrid_map = folium.Map(location=[40.4168, -3.7038], zoom_start=12)

# Iterar sobre el DataFrame y agregar cada punto al mapa
for index, row in ngdf.iterrows():
    location = [row['latitudes'][0], row['longitudes'][0]]
    tooltip = row['neighbourhood']
    folium.Marker(location=location, tooltip=tooltip).add_to(madrid_map)

# Mostrar el mapa
madrid_map
```

```python
import folium

# Coordenadas del centro de Madrid
madrid_coords = [40.416775, -3.703790]

# Cargar mapa de Madrid
m = folium.Map(location=madrid_coords, zoom_start=12)

# Agregar marcador en el centro de la ciudad
folium.Marker(
    location=madrid_coords,
    popup='Madrid',
    icon=folium.Icon(icon='cloud')
).add_to(m)

# Mostrar mapa interactivo
m





```

```python
m = folium.Map(location=[40.416775, -3.703790], zoom_start=12)

# Crear una capa para los marcadores de listings
listings_layer = folium.FeatureGroup(name='lis')

# Iterar a través del dataframe listings y agregar un marcador para cada ubicación
for i, row in lis.iterrows():
    lat = row['latitude']
    lon = row['longitude']
    popup = f"ID: {row['id']}\nBarrio: {row['neighbourhood_group_cleansed']}"
    folium.Marker([lat, lon], popup=popup).add_to(listings_layer)

# Agregar la capa listings_layer al mapa
m.add_child(listings_layer)

# Mostrar el mapa
m
```

```python
import plotly.express as px
fig = px.scatter_3d(lis, x='x', y='y', z='z', color='id', size_max=5)

# Agregue un título
fig.update_layout(title='Mapa interactivo de apartamentos Airbnb en Madrid')

# Muestre la figura
fig.show()
```

```python
fig = px.scatter_3d(lis, x='x', y='y', z='z', color='id', size_max=5, 
                    color_continuous_scale=px.colors.sequential.Plasma, 
                    range_color=(lis['price'].min(), lis['price'].max()),
                    color_continuous_midpoint=lis['price'].median())

# Agregue un título y etiquetas de eje
fig.update_layout(title='Mapa interactivo de apartamentos Airbnb en Madrid',
                  scene=dict(xaxis_title='Latitud', yaxis_title='Longitud', zaxis_title='Altitud'))

# Agregue una leyenda y ajuste el tamaño de la fuente
fig.update_traces(marker=dict(showscale=True, size=3),
                  selector=dict(mode='markers'))
fig.update_layout(coloraxis_colorbar=dict(title='price', len=0.75, y=0.45, yanchor='middle', 
                                          thicknessmode='pixels', thickness=20,
                                          tickfont=dict(size=10)))

# Muestre la figura
fig.show()
```

```python
import json
def parse_sentimientos(sentimiento):
    return json.loads(sentimiento.replace("'", "\""))

# Convertimos los datos en formato JSON a un diccionario de Python
rev['sentimiento_dict'] = rev['sentimiento'].apply(parse_sentimientos)

```

```python
# Creamos cuatro columnas nuevas y extraemos los valores correspondientes a cada clave en cada fila
rev['neg'] = rev['sentimiento_dict'].apply(lambda x: x['neg'])
rev['neu'] = rev['sentimiento_dict'].apply(lambda x: x['neu'])
rev['pos'] = rev['sentimiento_dict'].apply(lambda x: x['pos'])
rev['compound'] = rev['sentimiento_dict'].apply(lambda x: x['compound'])

# Eliminamos la columna original de sentimientos y la columna de diccionarios si ya no las necesitamos
```

```python
rev = rev.drop(['sentimiento', 'sentimiento_dict'], axis=1)
```

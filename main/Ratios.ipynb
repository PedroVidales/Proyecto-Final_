```python
pd.set_option('display.max_columns', None)
import pandas as pd
```

```python
df = pd.read_csv('ratios.csv')
lis = pd.read_csv('../madrid/listings_detailed/lis_ratios.csv')
```

```python
lis = lis.rename(columns={'id': 'listing_id'})
```

```python
merged_df = pd.merge(df, lis, on='listing_id', how='inner')
```

```python
if merged_df.isnull().values.any():
    print('Hay valores nulos en el archivo CSV')
else:
    print('No hay valores nulos en el archivo CSV')
```

```python
if merged_df.duplicated().any():
    print("Hay duplicados en el dataframe")
else:
    print("No hay duplicados en el dataframe")
```

```python
from tqdm import tqdm

# Definir una función para obtener el barrio a partir de la latitud y longitud
def get_barrio(lat, lng):
    # Obtener los resultados de geocodificación inversa para la ubicación dada
    results = gmaps.reverse_geocode((lat, lng))

    # Buscar el componente que tiene el tipo de "neighborhood" o "sublocality"
    for component in results[0]['address_components']:
        if 'neighborhood' in component['types'] or 'sublocality' in component['types']:
            return component['long_name']

    # Si no se encuentra el componente, devolver una cadena vacía
    return ''
```

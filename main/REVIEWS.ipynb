```python
import pandas as pd
import numpy as np 
import os 
from googletrans import Translator
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import warnings
warnings.filterwarnings('ignore')
```

```python
vader = pd.read_csv('../madrid/reviews_detailed/vader.csv')
sas = pd.read_csv('../madrid/reviews_detailed/sas.csv')
```

```python
vader.head()
```

```python
vader["negativo"] = vader.apply(lambda row: 1 if row["neg"] > row["pos"] else 0, axis=1)
vader["positivo"] = vader.apply(lambda row: 1 if row["pos"] > row["neg"] else 0, axis=1)
```

```python
vader.head()
```

```python
vader.drop(columns=["comments_short", "neg", "neu", "pos", "compound"], inplace=True)

```

```python
sas.drop(columns=["comments_short", "sentiment"], inplace=True)
```

```python
rev = pd.concat([vader, sas], ignore_index=True)
```
